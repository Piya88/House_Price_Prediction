# Problem of overfitting

From the above linear regression model output we can see that model works well on training data but it fails to predict test data. 
This because model tries to fit the train data so well that increased the model complexities and leads to overfitting problem.

One of the cause of overfitting is dimensionality issue of dataset. Dataset contains total 81 features which is difficult to for the regression model to provide appropriates weight for each feature.

Feature selection can reduce the problem of overfitting but it will again cause issue of under fitting due to loss of data.


### To reduce this problem we can use Regularization.

Regularization helps to overcome overfitting issue by reducing weights nearly equal to zero of unimportant features. This resolves dimensionality reduction issue and help to create generalized model which can also fit train & test data well.

#### There are 2 types of regularization: L1 & L2 regularization


## L2 Regularization: Ridge Regression


In L2 regularization we take summation of squared weights and add this to cost function.
It helps to assign appropriate weights for each features and that helps to reduce the complexity of model.

# Decision Tree Regressor


Decision tree regressor is used to predict continious target variable. Unlike decision tree we don't use entropy and gini-index to split the data. In Decision Tree Regressor model we uses Mean-Square-Error method to split the data.

### Decision tree Regressor model working

Step 1: At first step we will sort all the feature values and then we will take mean of first two values of each feature respectively. then we will split the data based on that mean value like : (x > M)or (x<= M) (x is datapoint & M is mean).

Step 2 : after splitting data based on mean of the first two values in to part 'A' (where x > M ) & part 'B' (where x <= M )we will calculate mean of target calues in part A and B respectively. This mean is nothing but the predicted value of (x > M)or (x<= M) and then we will calculate mean square error of predicted and original target value. we will note this MSE value for this split.


step 3 : continue the first step with the second value of feature and calculate MSE. In the end select split which gives less MSE. For higher dimentional data Select a split which gives lower MSE for each feature respectively and after that select a feature which splits gives lower MSE.

Step 4 : After getting root node, continue splitting further node into left ad right child node.

# Random Forest Regressor

Random forest regressor is supervised ensemble learning technique which uses bagging method to predict output.

Random forest creates multiple decision tree for each randomly selected samples with replacement call bootstrap technique and then calculates avarage of all the decision tree called aggregation output to make final prediction.


In Random Forest Regressor 
1] We will ranndomly take samples of data point and subset of features from given dataset 
2] based on this sample datapoints we will create decision tree regressor model and generate prediction for this regressor.
3] we will continue above steps to construct different decion tree regressors.
4] After that we will calculate avaerage of all decision tree regressor output as a final prediction of target variable.



#### Important parameters for decision tree regressor

n_estimators â€” the number of decision trees you will be running in the model
criterion â€” this variable allows you to select the criterion (loss function) used to determine model outcomes. We can select from loss functions such as mean squared error (MSE) and mean absolute error (MAE). The default value is MSE.
max_depth â€” this sets the maximum possible depth of each tree
max_features â€” the maximum number of features the model will consider when determining a split
bootstrap â€” the default value for this is True, meaning the model follows bootstrapping principles (defined earlier)
max_samples â€” This parameter assumes bootstrapping is set to True, if not, this parameter doesnâ€™t apply. In the case of True, this value sets the largest size of each sample for each tree.
Other important parameters are min_samples_split,

# Gradient boosting Machine algorithm for regressor

gradient boosting algorithm is uses ensemble method called boosting to make prediction for categorical and continious values.

Level 0:
1] It uses all recors and make predicyion for target cvariable. thn it calculates MSE by taking mean of all the predicted and observer target values.

level 1:
1] after that it create it's first estimator i.e. decision stump by randomly taking any feature as root node and split the data based on that feature.
         MSE = sum(OBSERVED - PRED)**2
2] now calculate mean and MSE of of splitted data and then add all the MSE value of each split that would be total MSE for that split. The resultant MSE of level 1 lesser than level 0 MSE and mean of each split is nothing but prediction of each split data point of 1st estimator

Level 2:
1] To create second estimator we will take reisdue's of level 1 (i.e. OBSERVED - PRED values for each record). We will create 2nd estimator by randomly taking any feature as root node to split target value. 
2] In this case our target alue is nothing but the residue of 1st estimator. we wiill split it based on root node and then calculate mean and MSE for each split.
3] Now, we need to subtract MSE and 2nd estimator split from mean of the 1st estimator.

And that will be fianl prediction of model with 2 level GBM estimator.

In GOSS:
We calculate gradients for each data points and then arrange in ascending order. Then select the higher gradient values and 2% lower gradient values. This helps model to get higher information gain using smaller dataset.

In EFB:
It compressses all the categorical encoded features into single feature, this reduces dimentionality of dataset and this have significant impact on overall accuracy of model.

With the comination of above 2 methods it fasten up overall training procedure.
